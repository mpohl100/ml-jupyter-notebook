```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
c = 299_792_458       # speed of light (m/s)
t = 1.0         # time (s)

# Angle range: 0 to 360 degrees
a_deg = np.linspace(0, 360, 2000)
a = np.deg2rad(a_deg)  # convert to radians

# Lambda_light equation
denominator = 2*c*t * (1.0 - 2*t*np.sin(a))
lambda_light = 1.0 / denominator

# Plot
plt.figure()
plt.plot(a_deg, lambda_light, label="λ_light")
plt.xlabel("Angle a (degrees)")
plt.ylabel("λ_light")
plt.title("λ_light vs Angle a")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
c = 299_792_458          # speed of light (m/s)
t = 1.0e-10                  # seconds
v = 0.9 * c              # electron velocity

# Angle range (degrees -> radians)
a_deg = np.linspace(0, 360, 2000)
a = np.deg2rad(a_deg)

# Term inside the square root
sqrt_arg = v**2 - (c * np.cos(a))**2

# Mask invalid (negative) values to avoid complex numbers
sqrt_term = np.where(sqrt_arg >= 0, np.sqrt(sqrt_arg), np.nan)

# Lambda expression
lambda_electron = 1.0 / ((c * np.sin(a) + sqrt_term) * 2 * t - v * t)

# Plot
plt.figure()
plt.plot(a_deg, lambda_electron)
plt.xlabel("Angle a (degrees)")
plt.ylabel("lambda_electron")
plt.title("lambda_electron vs angle a")
plt.grid(True)
plt.show()
```
